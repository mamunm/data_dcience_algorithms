From the above histogram of the data, we can infer several important characteristics of the data. 

1. The Serial No. is just the index of the data. We will just drop this column.
2. CGPA and GRE score are normal distributed feature, so is the target values (i.e., Chance of Admit)
3. LOR, Reseearch, SOP, TOEFL score, and University ranking are categorical feature. Fortunately, they are already in encoded format (i.e., they are numerical). So, we don't need to use label encoder or One hot encoder r similar transformer. 

```python
#Drop the Serial No. column
df = df.drop(['Serial No.'], axis=1)
```

```python
#Let us now plot the correlation between various features
fig,ax = plt.subplots(figsize=(8, 8))
sns.heatmap(df.corr(), ax=ax, annot=True, linewidths=0.05, fmt= '.2f',cmap="viridis")
plt.show()
```

From the correlation heatmap, we can see strong correlation between CGPA, GRE score and TOEFL score. A variable reduction mechanism might be very helpful to reduce the dimension of the model; however, we will focus on the whole dataset for now and later we will perform feature engineering to reduce the number of features.

```python
col = [i for i in df.columns if 'Chance' not in i]
X = df[col].values
y = df['Chance of Admit '].values
```

```python
data = preprocess_data(X, y)
data.clean_data()
X, y = data.get_data()
del data
```

```python
kernel_recipe = {'ConstantKernel' : [{'RBF' : [1.0,
                                                   {'length_scale' : 1.0}]},
                                      {'constant_value' : 1.0,
                                       'constant_value_bounds' : (3e-7, 3e7)}],
                 'WhiteKernel' : {'noise_level' : 0.1,
                                      'noise_level_bounds' : (1e-5, 1e5)}}
```

```python
r_state = [10, 20, 42, 80, 150, 200]
r_data = {rs : {} for rs in r_state}

for rs in r_state:
    print('    Working on random state: {}'.format(rs))
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.80, random_state=rs)

    MLGP = SKGP(X_train=X_train,
                X_test=X_test,
                y_train=y_train,
                y_test=y_test,
                kernel_recipe=kernel_recipe)

    MLGP.run_GP()
    r_data[rs] = MLGP.__dict__
    
if not os.path.exists('GP_data'):
    os.mkdir('GP_data')
    

np.save('GP_data/rsdata.npy', r_data)
LC = MLGP.plot_learning_curve()
LC.savefig('GP_data/learning_curve.png')
#PP = SKGP.parity_plot(data='train')
#PP.savefig('GP_data/parity_plot_train.png')
#PP = SKGP.parity_plot(data='test')
#PP.savefig('GP_data/parity_plot_test.png')
PP = MLGP.parity_plot(data='train', err_bar=True)
PP.savefig('GP_data/parity_plot_train_err_bar.png')
PP = MLGP.parity_plot(data='test', err_bar=True)
PP.savefig('GP_data/parity_plot_test_err_bar.png')
#PP = MLGP.parity_plot_fancy(data='train')
#PP.savefig('run_{0}/{0}_parity_plot_train.png'.format(ads))
#PP = MLGP.parity_plot_fancy(data='test')
#PP.savefig('run_{0}/{0}_parity_plot_test.png'.format(ads))
```

```python
data = np.load('GP_data/rsdata.npy')[()]
MAE_train = np.array([data[k]['MAE_train'] for k in data])
MAE_test = np.array([data[k]['MAE_test'] for k in data])
MSE_train = np.array([data[k]['MSE_train'] for k in data])
MSE_test = np.array([data[k]['MSE_test'] for k in data])

print('Mean absolute error for training and testing data:')
print('MAE_train: {}'.format(MAE_train))
print('MAE_train mean: {}'.format(MAE_train.mean()))
print('MAE_train std: {}'.format(MAE_train.std()))
print('MAE_test: {}'.format(MAE_test))
print('MAE_test mean: {}'.format(MAE_test.mean()))
print('MAE_test std: {}\n'.format(MAE_test.std()))

print('Mean squared error for training and testing data:')
print('MSE_train: {}'.format(MSE_train))
print('MSE_train mean: {}'.format(MSE_train.mean()))
print('MSE_train std: {}'.format(MSE_train.std()))
print('MSE_test: {}'.format(MSE_test))
print('MSE_test mean: {}'.format(MSE_test.mean()))
print('MSE_test std: {}\n'.format(MSE_test.std()))
```
